```python
#Chatboat with rule based responses
print("Chatbot: Hello! I am a simple chatbot. Type 'bye' to exit.")

name = "Tejaswini"

while True:
    user_input = input("You: ").lower()

    if user_input in ["hello", "hi"]:
        print("Chatbot: Hi there! 😊")

    elif "my name is" in user_input:
        name = user_input.split("my name is")[-1].strip().title()
        print(f"Chatbot: Nice to meet you, {name}!")

    elif "your name" in user_input:
        print("Chatbot: I am a simple chatbot.")

    elif "what is my name" in user_input:
        if name:
            print(f"Chatbot: Your name is {name}.")
        else:
            print("Chatbot: I don't know your name yet.")

    elif "how are you" in user_input:
        print("Chatbot: I'm doing well! Thanks for asking 😊")

    elif "help" in user_input:
        print("Chatbot: You can greet me, tell me your name, or ask questions.")

    elif user_input in ["bye", "exit"]:
        print("Chatbot: Goodbye! 👋")
        break

    else:
        print("Chatbot: Sorry, I don't understand that.")
```
